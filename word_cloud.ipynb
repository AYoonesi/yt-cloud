```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import os
```

```python
# extracting the text and replacing \n with space
video_id = 'ubV-8EYzTyc'
with open(f'tmp/{video_id}/{video_id}.txt') as seed:
    lines = seed.readlines()
seed = "".join(lines).replace('\n', ' ')
```

```python
# creating the word_cloud object
word_cloud = WordCloud(
    # background_color='#C3352E', collocations=False, random_state=3,
    width=1500, height=1500, 
    # contour_color='steelblue',
    # contour_width=3, contour_color='steelblue',
    background_color='white',
    max_words=80,
    stopwords=STOPWORDS , mask=np.array(Image.open('masks/pentagon.png'))
).generate(seed)
```

```python
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

```python
word_cloud.to_file(f'tmp/{video_id}/{video_id}.png')
```
