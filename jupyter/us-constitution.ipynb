```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import os
```

```python
with open('us-cons.txt') as seed:
    lines = seed.readlines()
seed = "".join(lines).replace('state', 'states').replace('\n', ' ').upper()
```

```python
# creating the word_cloud object
word_cloud = WordCloud(
    # background_color='#C3352E',
    # collocations=False, random_state=3,
     width=1700,
     height=3191, 
    # contour_color='steelblue',
    # contour_width=10, contour_color='white',
    background_color=None,
    mode='RGBA',
    # contour_width=0.5,
    # max_words=80,
    normalize_plurals=True,
    stopwords=STOPWORDS , mask=np.array(Image.open('lib.jpg'))
).generate(seed)
word_cloud.to_file('salam.png')
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

```python
import random
def grey_color_func(word, font_size, position, orientation, random_state=None,**kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

word_cloud.recolor(color_func=grey_color_func).to_file('grey.png')
```
